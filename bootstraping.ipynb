```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
data=pd.read_csv('cancer.csv')
```

```python
data.head()
```

```python
data['diagnosis'] = data['diagnosis'].apply(lambda x: '1' if x == 'M' else '0')
data = data.set_index('id')
del data['Unnamed: 32']
```

```python
data.head()
```

```python
print(data.groupby('diagnosis').size())
```

```python
Y = data['diagnosis'].values
X = data.drop('diagnosis', axis=1).values

X_train1, X_test1, Y_train1, Y_test1 = train_test_split (X, Y, test_size = 0.30, random_state=21)
```

```python
Y1= data['diagnosis'].values
X1= data.drop('diagnosis', axis=1).values

X_train2, X_test2, Y_train2, Y_test2 = train_test_split (X1, Y1, test_size = 0.70, random_state=21)
```

```python
a=X1[300:]
```

```python
x_train3=X1[::]
y_train3=Y1[::]
x_test3=X1[250:]
y_test3=Y1[250:]
```

```python
print(x_train3.size/30)
print(y_train3.size)
print(x_test3.size/30)
print(y_test3.size)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error
trainer1=LogisticRegression()

learner1=trainer1.fit(X_train1,Y_train1)

pd1=learner1.predict(X_test1)

acc1=accuracy_score(Y_test1,pd1)
print("accuracy of model=",acc1)

error1=mean_squared_error(Y_test1,pd1)

print("error =",np.sqrt(error1))

plt.plot(X_test1,Y_test1,'go-')
plt.plot(X_test1,pd1,'r*')
plt.xlabel("X_input")
plt.ylabel("Predicted_SVC")
plt.legend({"ya","yp"})
plt.show()
```

```python
trainer2=LogisticRegression()

learner2=trainer2.fit(X_train2,Y_train2)

pd2=learner2.predict(X_test2)

acc2=accuracy_score(Y_test2,pd2)
print("accuracy of model=",acc2)

error2=mean_squared_error(Y_test2,pd2)

print("error =",np.sqrt(error2))

plt.plot(X_test2,Y_test2,'go-')
plt.plot(X_test2,pd2,'r*')
plt.xlabel("X_input")
plt.ylabel("Predicted_SVC")
plt.legend({"ya","yp"})
plt.show()
```
